```python
import zipfile
import io

def decode_method(binary_data):
    bytes_object=io.BytesIO(binary_data)
    with zipfile.ZipFile(bytes_object, 'r') as zip_data:
        file_list=zip_data.namelist()
        srt_file_name=file_list[0]
        subtitle_content = zip_data.read( srt_file_name)
    return subtitle_content.decode('latin-1')
```

```python
text_data=decode_method(binary_data)
print(text_data)
```

```python
df['decoded_subtitle_content'] = df['content'].apply(decode_method)
```

```python
import re
```

```python
def content_clean(text_data):
  timestamp_pattern = re.compile(r'\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}')
  clean_data=re.sub(timestamp_pattern,'',text_data)
  clean_data=re.sub(r'\r\n',' ',clean_data)
  clean_data=re.sub(r'\d+',' ',clean_data)
  return clean_data
```

```python
clean_data=content_clean(text_data)
print(clean_data)
```

```python
df['subtitle_clean'] = df['decoded_subtitle_content'].apply(content_clean)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import whisper
model = whisper.load_model("base")
```

```python
def subtitle_finder(audio_path,top_k):
  result = model.transcribe(audio_path)
  query=result["text"]
  query_tf_idf= tfidf_vect.transform([query])
  cosine_score=cosine_similarity(query_tf_idf,tfidf_array)[0]
  cosine_score=list(enumerate(cosine_score))
  cosine_score_des=sorted(cosine_score,key=lambda x:x[1],reverse=True)
  sub_df=pd.DataFrame(columns=['num','name','Movie_Pred_Prob'])
  for score in cosine_score_des[:top_k]:
    sub_df.loc[len(sub_df)+1]=[df.iloc[score[0]]['num'],df.iloc[score[0]]['name'],round((score[1])*100)]
  sub_df.reset_index(inplace=True)
  return  sub_df
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings
```

```python
langchain_embed_model = SentenceTransformerEmbeddings()
```

```python
from langchain.vectorstores import Chroma

db1 = Chroma(collection_name="vector_database",
            embedding_function=langchain_embed_model,
            persist_directory="./chroma_db_")
```

```python
db1.add_documents(chunks)
```

```python
def subtitle_finder(audio_path,top_k):
  result = model.transcribe(audio_path)
  query=result["text"]
  query_vector = langchain_embed_model.embed_query(query)
  relevant_chunks = db1.similarity_search_by_vector(embedding=query_vector, k=top_k)
  sub_df=pd.DataFrame(columns=['num','name'])
  for doc in relevant_chunks:
    sub_df.loc[len(sub_df)+1]=[doc.metadata['num'],doc.metadata['name']]
  sub_df.reset_index(inplace=True)
  return  sub_df
```

```python
audio_path='the_message_5min.wav'
top_k=2
sub_df=subtitle_finder(audio_path,top_k)
sub_df
```
